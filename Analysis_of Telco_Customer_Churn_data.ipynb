### Creating Cohorts based on Tenure

**Let's begin by treating each unique tenure length as its own cohort.**

```python
tenura_Churn_rate = []
for i in range(1,73):
    rate = len(df[(df['tenure']==i) & (df['Churn'] == 'Yes')])/len(df[df['tenure']==i])*100
    tenura_Churn_rate.append(rate)
```

```python
tenura_Churn_rate
```

```python
ser_tenura = pd.Series(data=tenura_Churn_rate,index=range(1,73),name='tenura')
ser_tenura
```

**Now that I have Churn Rate per tenure group 1-72 months, create a plot showing churn rate per months of tenure.**

```python
plt.figure(figsize=(10,4),dpi=150)
plt.plot(ser_tenura.index, ser_tenura.values) 
plt.xlabel('tenure')
plt.ylabel('Churn percentage');
```

**You should have cohorts 1-72 months with a general trend of the longer the tenure of the cohort, the less of a churn rate. This makes sense as you are less likely to stop service the longer you've had it.**

### Broader Cohort Groups
**Based on the tenure column values, create a new column called Tenure Cohort that creates 4 separate categories:**
   * '0-12 Months'
   * '12-24 Months'
   * '24-48 Months'
   * 'Over 48 Months'    

```python
df['tenure']
```

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return 'Over 48 Months'
```

```python
df['Tenure_Cohort'] = df['tenure'].apply(cohort)
```

```python
df[['tenure','Tenure_Cohort']].head()
```

**Create a scatterplot of Total Charges versus Monthly Charts,colored by Tenure Cohort**

```python
plt.figure(figsize=(10,5),dpi=150)
sns.scatterplot(df,x='MonthlyCharges',y='TotalCharges',hue='Tenure_Cohort',alpha=0.5)
```

**Create a count plot showing the churn count per cohort.**

```python
plt.figure(figsize=(10,4),dpi=150)
sns.countplot(data=df,x=df['Tenure_Cohort'],hue='Churn')
```

# Predictive Modeling

**Let's explore some different tree based methods: A Single Decision Tree, Random Forest, XG Boosting.**

## Single Decision Tree

As proprocessing the data, Let's do a couple of tasks below  
ãƒ»drop ID column  
ãƒ»change the target column to 1,0  
ãƒ»get dummyvariables

```python
df['Churn'] = df['Churn'].map({'No':0,'Yes':1})
```

```python
X = pd.get_dummies(df.drop(['customerID','Churn'],axis=1),drop_first=True)
```

```python
X
```

```python
y = df['Churn']
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

**Decision Tree Perfomance.**

Let's take a look at what the best alpha for this model is 

```python
tree_model = DecisionTreeClassifier(random_state=101)
path = tree_model.cost_complexity_pruning_path(X_train, y_train)
eff_alphas, impurities = path.ccp_alphas, path.impurities
```

```python
tree_models = []
for eff_alpha in eff_alphas:
    model = DecisionTreeClassifier(random_state=101, ccp_alpha=eff_alpha)
    model.fit(X_train, y_train)
    tree_models.append(model)
```

```python
train_scores = [model.score(X_train, y_train) for model in tree_models]
test_scores = [model.score(X_test, y_test) for model in tree_models]
 
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.plot(eff_alphas, train_scores, marker="o", label="train", drawstyle="steps-post")
ax.plot(eff_alphas, test_scores, marker="o", label="test", drawstyle="steps-post")
ax.legend();
```

**the graph I showed above indicates alpha=0.015 is the best for this model. Let's see the accuracy when alpha is 0.015**

```python
best_tree_model = DecisionTreeClassifier(ccp_alpha=0.015)
best_tree_model.fit(X_train,y_train)
```

```python
tree_pred = best_tree_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
print(classification_report(y_test,tree_pred))
```

```python
ConfusionMatrixDisplay(confusion_matrix(y_test,tree_pred)).plot()
```

**This model does not seem to work properly to predict customer Churn.**  
**What is the important features in this model?**

```python
pd.DataFrame(index=X.columns,data=best_tree_model.feature_importances_,columns=['Feature Importance'])
```

```python
plot_tree(best_tree_model, filled=True, feature_names=X.columns);
```
